```python
from IPython.display import clear_output

def display_board(defaultboard, board):
    clear_output()
    print(' '+defaultboard[7]+' '+'|'+' '+defaultboard[8]+' '+'|'+' '+defaultboard[9]+' '+'    '+' '+board[7]+' '+'|'+' '+board[8]+' '+'|'+' '+board[9]+' ')
    print('---|---|---'+'    '+'---|---|---')
    print(' '+defaultboard[4]+' '+'|'+' '+defaultboard[5]+' '+'|'+' '+defaultboard[6]+' '+'    '+' '+board[4]+' '+'|'+' '+board[5]+' '+'|'+' '+board[6]+' ')
    print('---|---|---'+'    '+'---|---|---')
    print(' '+defaultboard[1]+' '+'|'+' '+defaultboard[2]+' '+'|'+' '+defaultboard[3]+' '+'    '+' '+board[1]+' '+'|'+' '+board[2]+' '+'|'+' '+board[3]+' ')
    
    
```

```python
def player_input():
    
    marker = ''
    
    while marker!='X' and marker!='O':
        marker = input('Player 1, Choose between X and O : ').upper()
        marker1 = marker
        
    if marker1 == 'X':
        return ('X','O')
    else:
        return ('O','X')
    '''
    OUTPUT : ( player1_marker, player2_marker )
    '''
```

```python
def place_marker(board, marker, position):
    
        
    board[position] = marker
```

```python
def win_check(board, mark):
    
    #ROWS CHECK
    return ((board[1]==board[2]==board[3]==mark) or 
    (board[4]==board[5]==board[6]==mark) or
    (board[7]==board[8]==board[9]==mark) or
    
    #COLS CHECK
    (board[1]==board[4]==board[7]==mark) or
    (board[2]==board[5]==board[8]==mark) or
    (board[3]==board[6]==board[9]==mark) or
       
    #DIAGS CHECK
    (board[1]==board[5]==board[9]==mark) or
    (board[3]==board[5]==board[7]==mark))
    
```

```python
import random

def choose_first():
    
    ch = random.randint(0,1)
    if ch == 0:
        print('Player 1')
    else:
        print('Player 2')
```

```python
def space_check(board, position):
    
    return board[position]==' '
```

```python
def full_board_check(board):
    
    for i in range(1,10):
        if space_check(board,i):
            return False
    
    return True
```

```python
def player_choice(board):
    
    choice = 0
    
    while choice not in range(1,10) or not space_check(board,choice):
        choice = int(input('Choose your next position : '))
    
    return choice
```

```python
def replay():
    
    replay = ''
    while replay not in ['Y','N']:
        replay = input('Play again??').upper()
        
    return replay == 'Y'
```

```python
print('Welcome to TIC-TAC-TOE !!!')

while True:
    
    ##PLAY THE GAME
    
    ##SET EVERYTHING UP( BOARD, WHOS FIRST, CHOOSE MARKERS)
    
    test_board = ['#','1','2','3','4','5','6','7','8','9']
    the_board = [' ']*10
    
    player1_marker,player2_marker = player_input()
    
    turn = choose_first()
    print(turn )
    
    play_game = input('Ready to play (Y/N) : ').upper()
    if play_game == 'Y':
        game_on = True
    else:
        game_on = False
        
    
    ##GAME PLAY
    
    while game_on:
        
        if turn == 'Player 1':
            display_board(test_board,the_board)
            
            #choose position
            position = player_choice(the_board)
            
            #place marker
            place_marker(the_board, player1_marker, position)
            
            #check if won
            if win_check(the_board,player1_marker):
                display_board(test_board,the_board)
                print('Player 1 Wins!!')
                game_on = False
            
            #check if tie
            if full_board_check(the_board):
                display_board(test_board,the_board)
                print('Tieee!!!')
                game_on = False
            #no tie no win ? nxt players turn
            else: 
                turn = 'Player 2'
    
    ##PLAYER 2 TURN    
        else:
            display_board(test_board,the_board)
            
            #choose position
            position = player_choice(the_board)
            
            #place marker
            place_marker(the_board, player2_marker, position)
            
            #check if won
            if win_check(the_board,player2_marker):
                display_board(test_board,the_board)
                print('Player 2 Wins!!')
                game_on = False
            
            #check if tie
            if full_board_check(the_board):
                display_board(test_board,the_board)
                print('Tieee!!!')
                game_on = False
            #no tie no win ? nxt players turn
            else: 
                turn = 'Player 1'
    
    
    
    
    if not replay():
        break
    
```
